```python
df = pd.read_csv('AtrasoCartaoCredito.csv')
```

```python
df.columns
```

#### Selecione as colunas `Target` e `Sexo`

```python
sqldf('''
    SELECT Target, Sexo
    FROM df
''')
```

#### Selecione as colunas `Target` e `Sexo` com pessoas de idade acima de 30 anos

```python
df.columns
```

```python
sqldf('''
    SELECT Target, Sexo
    FROM df
    WHERE Idade > 30
''')
```

#### Selecione as colunas `Target` e `Sexo` com pessoas de idade entre 30 e 50 anos

```python
sqldf('''
    SELECT Target, Sexo
    FROM df
    WHERE Idade BETWEEN 30 AND 50
''')
```

#### Selecione pessoas da região nordeste

```python
sqldf('''
    SELECT *
    FROM df
    WHERE RegiaodoPais = 'Região Nordeste'
''')
```

```python
sqldf('''
    SELECT *
    FROM df
    WHERE RegiaodoPais LIKE '%nord%'
''')
```

#### Selecione pessoas das regiões nordeste e sudeste

```python
sqldf('''
    SELECT *
    FROM df
    WHERE RegiaodoPais IN ('Região Nordeste', 'Região Sudeste')
''')
```

#### Selecione pessoas de todas regiões exceto sul

```python
sqldf('''
    SELECT *
    FROM df
    WHERE RegiaodoPais <> 'Região Sul'
''')
```

```python
sqldf('''
    SELECT *
    FROM df
    WHERE RegiaodoPais not in ('Região Sul')
''')
```

#### Selecione pessoas de MG e SP, e com idade acima de 30 anos

```python
sqldf('''
    SELECT *
    FROM df
    WHERE UF IN ('MG', 'SP') AND Idade > 30
''')
```

#### Selecione pessoas de sexo `homem`, residentes em  RJ e com idade abaixo de 50 anos

```python
sqldf('''
    SELECT *
    FROM df
    WHERE Sexo = 'homem' AND
            UF = 'RJ' AND
            Idade < 30
''')
```
